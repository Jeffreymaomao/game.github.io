```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
plt.rcParams["text.usetex"] = True
plt.rcParams["figure.dpi"] = 300
plt.rcParams["figure.figsize"] = (12,8)
plt.rcParams['lines.linewidth'] = 0.8

from GR_lens import GR_lens
Rs = 1
```

```python
#------------------------------------------------
bs = np.arange(1, 5, 0.2)
#------------------------------------------------
fig, ax = plt.subplots(1,1)
colors = plt.cm.jet(np.arange(0, 1, 1/len(bs)))
for b, color in zip(bs,colors):
    x,y = GR_lens(Rs=Rs, U0=[0,1/b/Rs])
    plt.plot(x, y, color=color, label="$b=%.2fR_s$"%b)
ax.add_artist(plt.Circle((0,0), Rs, color='black',zorder=10)) 
ax.axis('equal')
ax.set_xlim(-5*Rs,5*Rs)
ax.set_ylim(-5*Rs,5*Rs)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.legend()
plt.show()
```

```python
#------------------------------------------------
angles = np.arange(0, 1, 0.1)
b = 10
#------------------------------------------------
fig, ax = plt.subplots(1,1)
colors = plt.cm.jet(np.arange(0, 1, 1/len(angles)))
for angle, color in zip(angles,colors):
    x,y = GR_lens(Rs=Rs, U0=[angle, 1/b/Rs])
    plt.plot(x, y, color=color, label="$\\theta=%.2f^\\circ$"%(angle*180/np.pi))
ax.add_artist(plt.Circle((0,0), Rs, color='black',zorder=10)) 
ax.axis('equal')
ax.set_xlim(-5*Rs,5*Rs)
ax.set_ylim(0, 1.2*b*Rs)
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.legend()
plt.show()
```
